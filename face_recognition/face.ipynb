```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline
```

```python
#read Image 
image = cv2.imread('chris.jpg')
#convert to grayscale
gray = cv2.cvtColor(image , cv2.COLOR_BGR2GRAY)
#APPLY harr cascade
harr = cv2.CascadeClassifier('model/haarcascade_frontalface_default.xml')
face = harr.detectMultiScale(gray,1.3,5)#(image ,scale,minimum neighbours)
print(face)
```

```python
cv2.rectangle(image ,(154,94),(154+261,94+261),(0,255,0),2)
plt.imshow(image)
```
