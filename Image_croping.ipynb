```python
gray_cv = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(gray_cv,cmap='gray')
```

```python
#Load Harcascadeclassifier

haar = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
faces = haar.detectMultiScale(gray_cv,1.5,2)
faces
```

```python
for x,y,w,h in faces:
    cv2.rectangle(gray_cv,(x,y),(x+w,y+h),(0,122,90),2)
    plt.imshow(gray_cv,cmap='gray')
```

```python
crop_image = gray_cv[y:y+h,x:x+w]
plt.imshow(crop_image)
```

```python
cv2.imwrite('p1.png',crop_image)
```

```python
#Applying the techniques from reading to cropping and storing in all images
harr = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
def extract_images(path,gender,i):
    img = cv2.imread(path)
    gray_img = cv2.cvtColor(img , cv2.COLOR_BGR2GRAY)
    face = harr.detectMultiScale(gray_img,1.5,5)
    for x,y,w,h in face:
        crop_image = gray_img[y:y+h,x:x+w]
        if gender == 'male':
            cv2.imwrite('data/male_crop/{}_{}.png'.format(gender,i),crop_image)
        else:
            cv2.imwrite('data/female_crop/{}_{}.png'.format(gender,i),crop_image)
```

```python
extract_images(femalepath[1],'female',1)
```

```python
for i,path in enumerate(femalepath):
    try:
        extract_images(path,'female',i)
        print('Info : {}/{} processed succesfully'.format(i,len(femalepath)))
    except:
        
        print('Info : {}/{} cannot be processed'.format(i,len(femalepath)))
```

```python
for i,path in enumerate(malepath):
    try:
        extract_images(path,'male',i)
        print('Info : {}/{} processed succesfully'.format(i,len(femalepath)))
    except:
        
        print('Info : {}/{} cannot be processed'.format(i,len(femalepath)))
```
